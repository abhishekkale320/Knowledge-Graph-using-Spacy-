```python
import networkx as nx
import plotly.graph_objects as go
import random

# Generate a 3D position for each node using a random layout
def generate_3d_positions(G):
    pos = {node: (random.uniform(-1, 1), random.uniform(-1, 1), random.uniform(-1, 1)) for node in G.nodes}
    return pos

# Extract all nodes and relationships as before (assuming this is done)
# Add nodes and edges
for entity, label in entities_list:
    G.add_node(entity, label=label)

for subject, relationship, object_ in relationships_list:
    G.add_edge(subject, object_, relationship=relationship)

# Generate 3D positions
pos = generate_3d_positions(G)

# Create a 3D network graph using Plotly
edge_x = []
edge_y = []
edge_z = []

for edge in G.edges():
    x0, y0, z0 = pos[edge[0]]
    x1, y1, z1 = pos[edge[1]]
    edge_x += [x0, x1, None]
    edge_y += [y0, y1, None]
    edge_z += [z0, z1, None]

edge_trace = go.Scatter3d(
    x=edge_x, y=edge_y, z=edge_z,
    line=dict(width=2, color='gray'),
    hoverinfo='none',
    mode='lines'
)

node_x = []
node_y = []
node_z = []
node_text = []

for node in G.nodes():
    x, y, z = pos[node]
    node_x.append(x)
    node_y.append(y)
    node_z.append(z)
    node_text.append(f"Node: {node}, Label: {G.nodes[node].get('label', 'No Label')}")

node_trace = go.Scatter3d(
    x=node_x, y=node_y, z=node_z,
    mode='markers+text',
    marker=dict(symbol='circle', size=10, color='lightblue', line=dict(width=2)),
    text=node_text,
    hoverinfo='text'
)

layout = go.Layout(
    title="3D Graph Visualization",
    showlegend=False,
    scene=dict(
        xaxis=dict(showbackground=False),
        yaxis=dict(showbackground=False),
        zaxis=dict(showbackground=False),
    )
)

fig = go.Figure(data=[edge_trace, node_trace], layout=layout)
fig.show()
```
